```python
dataset[["dataset", "frequency", "term"]] = dataset["dataset"].str.split("/", expand=True)
```

```python
pretty_names = {
    "jena_weather": "Jena Weather",
    "solar": "Solar",
    "bitbrains_fast_storage": "Bitbrains Fast Storage",
    "bitbrains_rnd": "Bitbrains RND",
    "bizitobs_l2c": "Bizitobs L2C",
    "bizitobs_application": "Bizitobs Application",
    "bizitobs_service": "Bizitobs Service",
}
```

```python
# organize by domain and dataset
dataset = dataset.sort_values(["domain", "dataset", "frequency"])
dataset = dataset.reset_index(drop=True)

dataset["dataset"] = dataset["dataset"].map(pretty_names)
dataset["frequency"] = dataset["frequency"].map({
    "5T": "5 min",
    "H": "1 hour",
    "D": "1 day",
    "W": "1 week",
    "10T": "10 min",
    "10S": "10 sec",
})

dataset.head()
```

```python
dataset.to_csv("results/datasets.csv", index=False)
```

```python
# raccolgi per domain, dataset, frequency
dataset = dataset.groupby(["domain", "dataset", "frequency"]).agg({
    "num_samples": "first",
    "num_variates": "first",
    "term": lambda x: ", ".join(sorted(set(x)))
}).rename(columns={
    "num_samples": "Num. Samples",
    "num_variates": "Num. Variates",
    "term": "Term"
}).reset_index()

dataset
```

```python
# plot ts
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime, timedelta
from matplotlib.ticker import MaxNLocator
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
import matplotlib.ticker as ticker


def plot_ts(df, title, x_label, y_label, legend_title, save_path):
    fig, ax = plt.subplots(figsize=(12, 6))
    for i, (name, group) in enumerate(df.groupby("dataset_name")):
        group = group.reset_index(drop=True)
        if "target" in group.columns:
            target = group["target"][0]
            if isinstance(target, pd.DataFrame):
                target = target.iloc[:, 0]
            ax.plot(target.index, target.values, label=name)
        else:
            ax.plot(group.index, group.values, label=name)

    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.legend(title=legend_title)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(save_path)
    plt.show()
    #plt.close()

# Plotting the time series data
plot_ts(solar_10T, "Benito's L2C 5T", "Time", "Value", "Dataset Name", "results/bizitobs_l2c_5T.png")
```
